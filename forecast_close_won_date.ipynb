```python
seed(1)
lstm_data = pd.DataFrame(columns=['date' , 'monthly_booking_amount' , 'x1','x2','x3','x4','y'])
```

```python
# grouped_dataframe = pd.read_csv('grouped_dataframe.csv')
# grouped_dataframe = pd.read_csv('test.csv')
grouped_dataframe.reset_index(inplace=True)
grouped_dataframe.shape
```

```python
lstm_data['date'] = grouped_dataframe['index']
lstm_data['monthly_booking_amount'] = grouped_dataframe['Booking Amount']
lstm_data['x1'] = grouped_dataframe['Booking Amount']
lstm_data['x2'] = grouped_dataframe['Booking Amount'].shift(-1)
lstm_data['x3'] = grouped_dataframe['Booking Amount'].shift(-2)
lstm_data['x4'] = grouped_dataframe['Booking Amount'].shift(-3)
# lstm_data['x5'] = grouped_dataframe['Booking Amount'].shift(-4)
lstm_data['y'] = grouped_dataframe['Booking Amount'].shift(-4)
```

```python
lstm_data.head()
```

```python
new_data = lstm_data.loc[: , 'x1':].dropna()
```

```python
new_data.tail()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler = StandardScaler()
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_df = pd.DataFrame(scaler.fit_transform(new_data) ,columns=['x1','x2','x3','x4','y'])
```

```python
len(lstm_data[105:123]["date"])
```

```python
x_train ,y_train = np.array(scaled_df.iloc[:105, :-1]) ,np.array(scaled_df.iloc[:105, -1])
x_test ,y_test = np.array(scaled_df.iloc[105: ,:-1]) , np.array(scaled_df.iloc[105: , -1])
```

```python
x_test.shape
```

```python
x_train_reshaped = x_train.reshape((x_train.shape[0],1,x_train.shape[1]))
y_train_reshaped = y_train.reshape((y_train.shape[0] ,1 , 1))
x_val_resaped =    x_test.reshape((x_test.shape[0], 1,x_test.shape[1]))
#batch_size =1
```

```python
# model = Sequential()
# model.add(LSTM(128, return_sequences=True))
# model.add(Activation('relu'))
# model.add(LSTM(128, return_sequences=True))
# model.add(Activation('relu'))
# model.add(LSTM(64, return_sequences=False))
# model.add(Activation('relu'))
# #model.add(BatchNormalization())
# #model.add(Dense(128))
# model.add(Dense(1,activation='linear'))
# model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
print(x_train_reshaped.shape)
print(x_val_resaped.shape)
```

```python
model = Sequential()

model.add(LSTM(96 ,input_shape = (1,4) , return_sequences=True))

model.add(Activation('relu'))

model.add(LSTM(96, return_sequences=False))

model.add(Activation('relu'))

model.add(Dense(1,activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
#model.fit(x_train, y_train, batch_size=16, epochs=120)
```

```python
# fit model
filename = 'model_1.h5'
checkpoint = ModelCheckpoint(filename, monitor='loss', verbose=1, save_best_only=True, mode='min')
history = model.fit(x_train_reshaped, np.array(y_train), batch_size=3, epochs=50, callbacks=[checkpoint], verbose=1)

# history = model.fit(x_train_reshaped, np.array(y_train), batch_size=12, epochs=100)
#history = model.fit(train_X, train_y, epochs=10, batch_size=20, , verbose=1, shuffle=False)
# summarize history for loss
plt.plot(history.history['loss'])
# plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
# model = load_model('model_1.h5')
y_pre = model.predict(x_val_resaped)
```

```python
forecast_pre['actual'] = new_data.iloc[105:, -1:].values
```

```python
error = mean_squared_error(forecast_pre['prediction'], forecast_pre['actual'])
print('Test MSE: %.3f' % error**0.5)
```
